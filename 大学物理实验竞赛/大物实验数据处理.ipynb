```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 设置中文字体
plt.rcParams['font.family'] = 'SimHei'  # 设置中文字体
plt.rcParams['axes.unicode_minus'] = False  # 防止负号显示为方块

# 假设 df_1 是你已经读取的 DataFrame
# 计算每一列的均值和 A 类不确定度，并打印输出
for column_name in df_1.columns:
    mean = df_1[column_name].mean()  # 计算均值
    std = df_1[column_name].std()    # 计算标准差
    u_a = std / np.sqrt(len(df_1[column_name]))  # 计算 A 类不确定度

    # 打印输出结果
    print(f'直流偏移量为 {column_name} V, 质心偏移量A类不确定度为 {u_a:.4f} mm, 均值为 {mean:.4f} mm')

# 定义要绘制的列名称
column_names = [-11, 3, 11]

# 创建一个 3 行 1 列的子图布局
plt.figure(figsize=(10, 15))  # 设置图形大小

for i, column_name in enumerate(column_names):
    # 计算均值和不确定度
    mean = df_1[column_name].mean()  # 计算均值
    std = df_1[column_name].std()    # 计算标准差
    u_a = std / np.sqrt(len(df_1[column_name]))  # 计算 A 类不确定度

    # 在子图中绘制直方图
    plt.subplot(3, 1, i + 1)  # 创建 3x1 的子图，选择第 i+1 个位置
    plt.hist(df_1[column_name], bins=30, color='skyblue', edgecolor='black', alpha=0.7)
    
    # 绘制均值和不确定度线
    plt.axvline(mean, color='red', linestyle='dashed', linewidth=1, label=f'均值 = {mean:.4f}')
    plt.axvline(mean + u_a, color='green', linestyle='dashed', linewidth=1, label=f'A类不确定度上限 = {mean + u_a:.4f}')
    plt.axvline(mean - u_a, color='green', linestyle='dashed', linewidth=1, label=f'A类不确定度下限 = {mean - u_a:.4f}')

    # 添加标题和标签
    plt.title(f'直方图（直流偏移量 {column_name} V）')
    plt.xlabel('值 (mm)')
    plt.ylabel('频数')
    plt.legend()
    plt.grid()

# 显示所有的子图
plt.tight_layout()  # 自动调整子图参数
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

plt.rcParams['font.family'] = 'SimHei'  # 设置中文字体
plt.rcParams['axes.unicode_minus'] = False  # 防止负号显示为方块

# 假设 df_2 已经定义并包含你需要的数据
B = df_2['磁感应强度大小']
V = df_2['直流偏移量']

# 去掉 B 和 V 中存在缺失值的行
valid_data = df_2.dropna(subset=['磁感应强度大小', '直流偏移量'])

# 重新提取有效的数据
B_valid = valid_data['磁感应强度大小']
V_valid = valid_data['直流偏移量']

# 查看有效数据
print(B_valid)
print(V_valid)

# 绘制散点图
plt.scatter(V_valid, B_valid, label='数据点')  # 散点图
plt.xlabel('直流偏移量/V')  # X轴标签
plt.ylabel('磁感应强度大小/T')  # Y轴标签
plt.title('磁感应强度与直流偏移量点线图')  # 标题

# 线性拟合
slope, intercept, r_value, p_value, std_err = linregress(V_valid, B_valid)

# 生成拟合线
fit_line = slope * V_valid + intercept

# 绘制拟合线
plt.plot(V_valid, fit_line, color='red', label='拟合线: y = {:.7f}x '.format(slope))

# 添加图例
plt.legend()

# 显示图形
plt.show()

# 输出拟合效果分析
print(f'拟合斜率: {slope}')
print(f'拟合截距: {intercept}')
print(f'R-squared: {r_value**2}')
print(f'p-value: {p_value}')
print(f'标准误差: {std_err}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义常量
lamda = 532e-9  # 波长 (m)
z = 0.8  # 传播距离 (m)
k = 2 * np.pi / lamda  # 波数 (m^-1)
V = 50  # 速度 (m/s)
l = 0.05  # 计算用的长度 (m)
w0 = 20e-6  # 初始光斑半径 (m)
R0 = w0**2 * k / 2  # 曲率半径

# 计算 eta 的具体值
r_s = 1  
r_p = 2  
eta = r_s / r_p

# 计算 delta_y_H 和 delta_y_V
delta_y_H = (1 + eta) / k
delta_y_V = (1 + 1 / eta) / k

# 计算 delta_y 的两个值
def Delta(psi):
    Delta_y = z / R0 * 2 * psi * eta * delta_y_H / (
        (1 + psi**2 * eta**2) * np.exp(k * delta_y_H**2 / R0) + psi**2 * eta**2 - 1
    )
    return Delta_y

def Delta_(psi, sigma):
    Delta_y = sigma / l / V * psi
    return Delta_y

# 定义 sigma 的值
sigma = 1.326  

# 牛顿法的目标函数和它的导数
def f(psi):
    return Delta(psi) - Delta_(psi, sigma)

def df(psi):
    h = 1e-6  # 小的增量，用于数值求导
    return (f(psi + h) - f(psi - h)) / (2 * h)

# 牛顿法的实现
def newton_method(initial_psi, max_iter=100, tol=1e-6):
    psi = initial_psi
    for _ in range(max_iter):
        psi_new = psi - f(psi) / df(psi)  # 更新 psi
        if abs(psi_new - psi) < tol:  # 判断是否收敛
            return psi_new
        psi = psi_new
    print(f"Warning: Newton's method did not converge for initial guess {initial_psi}")
    return None  # 如果未收敛则返回 None

# 在 0 到 0.01 的范围内寻找合适的初始值
initial_guesses = np.linspace(0, 0.001, 10)  
solutions = []

for initial_psi in initial_guesses:
    root = newton_method(initial_psi)
    if root is not None and root not in solutions:
        solutions.append(root)

# 输出结果
print("使用牛顿法找到的根:", solutions)

# 可视化结果
B_values = np.linspace(0, 0.003, 10000)
psi_values = l * V * B_values
delta_y_approx_values = [Delta_(psi, sigma) for psi in psi_values]
delta_y_computed_values = [Delta(psi) for psi in psi_values]

plt.figure(figsize=(12, 6))
plt.plot(B_values, delta_y_approx_values, label='近似值 Δy_approx', color='blue')
plt.plot(B_values, delta_y_computed_values, label='非近似值 Δy', color='red')


plt.axvline(x=solutions[-1], color='green', linestyle='--', label=f'$B_max$：{solutions[-1]:.2e} T')

plt.xlabel('B/T')
plt.ylabel('Δy 值')
plt.title(' Δy 值与 B 的关系')
plt.legend()
plt.grid()
plt.show()
```
